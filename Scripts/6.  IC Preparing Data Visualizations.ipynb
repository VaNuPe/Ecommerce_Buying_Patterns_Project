## Data quality and consistency checks

```python
# Check for mixed types

for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

```python
# Fixing the inconsistency

customers['First Name'] = customers['First Name'].astype('str')
```

```python
# 2nd Check for mixed types

for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

## Combine the data

```python
# Changing 'user_id' data type to str

ords_prods_merge['user_id'] = ords_prods_merge['user_id'].astype('str')
```

```python
# To avoid a ValueError (Cannot use name of an existing column for indicator column), we need to delete the _merge column
# I will drop as well other columns we don't need anymore, like 'max_order'

ords_prods_merge = ords_prods_merge.drop(columns = ['_merge', 'max_order', 'average_spending', 'regularity'])
```

```python
# Merging the data

orders_products_all = ords_prods_merge.merge(customers, on = 'user_id', indicator = True)
```
